```python
data.SalaryUSD = data.SalaryUSD.apply(lambda x: float(str(x).replace(',','').replace('$','').replace(' ','')))
```

```python
#encode the data
le = preprocessing.LabelEncoder()
```

```python
# data.Country = le.fit_transform(data.Country)
data.EmploymentStatus = le.fit_transform(data.EmploymentStatus)
# data.JobTitle = le.fit_transform(data.JobTitle)            
# data.ManageStaff = le.fit_transform(data.ManageStaff)
data.YearsWithThisTypeOfJob = le.fit_transform(data.YearsWithThisTypeOfJob)
data.Education = le.fit_transform(data.Education)
# data.EducationIsComputerRelated = le.fit_transform(data.EducationIsComputerRelated)
data.Certifications = le.fit_transform(data.Certifications)
data.HoursWorkedPerWeek = le.fit_transform(data.HoursWorkedPerWeek)
# data.EmploymentSector = le.fit_transform(data.EmploymentSector)
data.LookingForAnotherJob = le.fit_transform(data.LookingForAnotherJob)
```
